# Questions based on Visualization  :

 Which season consisted of the highest number of matches ever played ?


```python

matches_per_season=ps.groupby('season').size()
max_matches_by_season = matches_per_season.idxmax()
max_match=matches_per_season.max()

max_matches_by_season , max_match
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(matches_per_season.index, matches_per_season.values, color='skyblue')
for bar in bars:
    if bar.get_x() == max_matches_by_season:
        bar.set_color('orange')

plt.title('Matches Played per Season', fontsize=16)
plt.xlabel('Season', fontsize=14)
plt.ylabel('Number of Matches', fontsize=14)

plt.text(
    max_matches_by_season,
    max_match + 1,
    f'{max_match}',
    ha='center',
    fontsize=12,
    color='black'
)

plt.xticks(matches_per_season.index, rotation=45)
plt.tight_layout()
plt.show()
```

 The total runs by fours hit and the total number of fours hit by each team

```python

total_four=ps1[ps1['batsman_runs']==4]
total_run_by_four=total_four.groupby('batting_team')['batsman_runs'].sum().reset_index()
four_hitting=total_four.groupby('batting_team').size().reset_index(name='total_fours')
four_state=pd.merge(total_run_by_four,four_hitting,on='batting_team')

four_state
```

The fours hit by players


```python

four_hitting_by_player = ps1[ps1['batsman_runs'] == 4]
fours_count=four_hitting_by_player.groupby('batsman').size().reset_index(name='total_fours')
fours_by_player_count=fours_count.sort_values(by='total_fours' , ascending=False)

fours_by_player_count
```

```python
players = fours_by_player_count['batsman'].head(10)  # Top 10 players
fours = fours_by_player_count['total_fours'].head(10)
plt.figure(figsize=(10, 6))
for i in range(len(players)):
    plt.bar(players.iloc[i], fours.iloc[i])
plt.xlabel('Players')
plt.ylabel('Total Fours')
plt.title('Top 10 Players with Most Fours')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

 The number of fours hit in each season


```python

fours=ps1[ps1['batsman_runs'] == 4]
season_fours=fours.merge(ps[['id', 'season']], left_on='match_id', right_on='id')
each_season_four=season_fours.groupby('season').size().reset_index(name='total_fours')
fours_by_season_state=each_season_four.sort_values(by='season')

fours_by_season_state
```

 The total runs by the sixes hit and the number of sixes hit by each team


```python

sixes = ps1[ps1['batsman_runs'] == 6]
sixes_count=sixes.groupby('batting_team').size().reset_index(name='total_sixes')
sixes_count['total_runs_by_six']=sixes_count['total_sixes']*6
six_by_team=sixes_count.sort_values(by='total_sixes',ascending=False)
six_by_team
```

The number of sixes hit in each season


```python
sixes = ps1[ps1['batsman_runs'] == 6]
season_sixes=sixes.merge(ps[['id', 'season']], left_on='match_id', right_on='id')
each_season_six=season_sixes.groupby('season').size().reset_index(name='total_sixes')
six_state=each_season_six.sort_values(by='season')

six_state
```
